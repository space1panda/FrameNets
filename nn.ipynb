```python
def sigmoid(x):
    return 1/(1+ np.exp(-x))

def sigmoid_p(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
# scatter data

plt.axis([0,6,0,6])

plt.grid()

for i in range(len(data)):
    point = data[i]
    color = 'r'
    if point[2] ==0:
        color = 'b'
    plt.scatter(point[0], point[1], c=color)
```
